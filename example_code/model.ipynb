```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
```

# Before modeling

```python
df = pd.read_csv('./FS_2y_before_Bankruptcy_train_preprocessed.csv').drop('Unnamed: 0',1)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns = ['class']),df['class'], test_size = 0.3)
```

# The models by 'sklearn'

- XGboost
- LightGBM
- RandomForest
- Catboost
- Logistic regression

# Logistic Regression

```python
model = LogisticRegression(random_state=2021, max_iter=1000, solver='saga')
```

```python
stf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 2021)
```

```python
score = cross_val_score(model, X_train, y_train, cv = stf, scoring = 'f1_micro')
score.mean()
```

```python
model.fit(X_train,y_train)
```

```python
pred_y = model.predict(X_test)
```

```python
print(classification_report(y_test, pred_y))
print("정상으로 예측 중 위험인 확률 : %.3f" % np.mean(y_test[pred_y == 0]))
```

# RandomForest

```python
model = RandomForestClassifier(random_state =1, n_estimators = 500)
```

```python
stf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 1)
```

```python
score = cross_val_score(model, X_train, y_train, cv = stf, scoring = 'f1_micro')
```

```python
score.mean()
```

```python
model.fit(X_train,y_train)
```

```python
pred_y = model.predict(X_test)
```

```python
print(classification_report(y_test, pred_y))
print("정상으로 예측 중 위험인 확률 : %.3f" % np.mean(y_test[pred_y == 0]))
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import precision_score, recall_score, f1_score

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
```

# Before modeling

```python
df = pd.read_csv('./FS_2y_before_Bankruptcy_train_preprocessed.csv')
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns = ['class']),df['class'], test_size = 0.3)
```

# The models by 'sklearn'

- XGboost
- LightGBM
- RandomForest
- Catboost
- Logistic regression

# Logistic Regression

```python
model = LogisticRegression(random_state=1)
```

```python
stf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 1)
```

```python
score = cross_val_score(model, X_train, y_train, cv = stf, scoring = 'f1_micro')
score.mean()
```

```python
model.fit(X_train,y_train)
```

```python
pred_y = model.predict(X_test)
```

```python
print("F1 : %.3f" % f1_score(y_test, pred_y, average = 'micro'))
```

# RandomForest

```python
model = RandomForestClassifier(random_state =1, n_estimators = 500)
```

```python
stf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 1)
```

```python
score = cross_val_score(model, X_train, y_train, cv = stf, scoring = 'f1_micro')
```

```python
score.mean()
```

```python
model.fit(X_train,y_train)
```

```python
pred_y = model.predict(X_test)
```

```python
print("F1 : %.3f" % f1_score(y_test, pred_y, average = 'micro'))
```
